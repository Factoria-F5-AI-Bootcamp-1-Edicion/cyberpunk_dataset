# Hipótesis gráfico años/publicaciones:
## el gráfico se pudiese explicar a través de eventos culturales del siglo XX y XXI. Antes de los años 60 no se tienen casi publicaciones en el género, intuimos que tiene que ver con la segunda guerra mundial y la crisis mundial derivada. Luego entre los años 60 y 70, se nota un pequeño pico de publicaciones y creemos que se debe al movimiento de contra cultura en Estados Unidos, el cual desafía el status quo, y este es un concepto muy presente en el género cyberpunk. A partir de los años 80 en adelante, notamos que el gráfico entra en una tendencia ascendente. Consideramos que se debe al lanzamiento de la pélicula "Blade Runner" del director Ridley Scott, la cual es uno de los ejemplos más clásicos del género Cyberpunk en la industria del entretenimiento. Al haberse convertido en una pelicula de culto, suponemos que influyó en la demanda de este género literario. En el año 2018 vemos el mayor pico del gráfico el cual coincide con la gran campaña de marketing para el videojuego cyberpunk 2077. Esta campaña, consideramos que tiene una relación directa con el aumento de la popularidad del género en la cultura popular. Por último vemos un bajón hacia el año 2020 el cúal es muy fácil intuir que se deba a la pandemia del covid 19.
